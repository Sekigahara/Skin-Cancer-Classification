```python
import cv2
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Get Every image path
file_path_benign = glob.glob("dataset_three_classes/train/benign/*")
file_path_semi = glob.glob("dataset_three_classes/train/semi/*")
file_path_malign = glob.glob("dataset_three_classes/train/malign/*")
```

```python
def load_image(list_path):
    images = []
    for idx, img_path in enumerate(list_path):
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        images.append(img)

    return np.array(images)

def display_images(images):
    for idx, img in enumerate(images):
        
        plt.subplot(1, 4, idx + 1)
        plt.imshow(img)
        plt.show()
```

```python
images_benign, images_semi, images_malign = load_image(file_path_benign), load_image(file_path_semi), load_image(file_path_malign)
```

```python
display_images(images_benign[:4])
```

```python
display_images(images_semi[:4])
```

```python
display_images(images_benign[:4])
```
